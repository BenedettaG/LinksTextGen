```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams['figure.dpi'] = 300
  
X = ['p=0.1','p=0.95','p=0.99']
selfBleu2 = [0.727260,0.568244, 0.579229]
selfBleu3 = [0.572027,0.359059, 0.366049]
selfBleu4 = [0.471779, 0.222479,0.224833]
selfBleu5 = [0.405355,0.148608, 0.151494]
  
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.15, selfBleu2, 0.1, label = 'Self-BLEU2')
plt.bar(X_axis - 0.05, selfBleu3, 0.1, label = 'Self-BLEU3')
plt.bar(X_axis + 0.05, selfBleu4, 0.1, label = 'Self-BLEU4')
plt.bar(X_axis + 0.15, selfBleu5, 0.1, label = 'Self-BLEU5')
  
plt.xticks(X_axis, X)
plt.title("Self-BLEU varying NUCLEUS parameter")
plt.legend()
plt.show()
```
